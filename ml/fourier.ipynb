```python
from utils import *
from scipy.signal import firwin
from scipy.fft import rfft, rfftfreq
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fir_filter_coef(ntaps, lo, hi, freq):
    coef = firwin(
        ntaps, [lo, hi], fs=freq, pass_zero=False, window="hamming", scale=False
    )
    return coef


def fir_filter(signal, coef):
    return np.convolve(signal, coef, "valid")


FREQ = 200
FILT_COEF = fir_filter_coef(10, 5, 40, FREQ)
```

```python
signal_2 = read_signal(1, 1, 2)
filt_signal_2 = fir_filter(signal_2, FILT_COEF)
ampl = np.abs(rfft(filt_signal_2))
freq = rfftfreq(len(filt_signal_2), 1 / 200)
_ = plt.plot(freq, ampl)
```
